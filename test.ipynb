```python
class Solution:
    def __init__(self):
        self.dic = {}
    def check(self, grid, row, col):
        if (row, col) in self.dic:
            return self.dic[(row,col)]
        if row >= len(grid) or col >= len(grid[0]):
            return 0
        if row == len(grid) - 1 and col == len(grid[0]) - 1:
            temp = grid[row][col]
            self.dic[(row,col)] = temp
            return temp
        if row == len(grid) - 1:
            temp = grid[row][col] + self.check(grid, row, col + 1)
            self.dic[(row,col)] = temp
            return temp
        if col == len(grid[0]) - 1:
            temp = grid[row][col] + self.check(grid, row + 1, col)
            self.dic[(row,col)] = temp
            return temp
        else:
            temp1 = grid[row][col] + self.check(grid, row + 1, col)
            temp2 = grid[row][col] + self.check(grid, row, col + 1)
            temp = min(temp1, temp2)
            self.dic[(row,col)] = temp
            return temp
    def minPathSum(self, grid: List[List[int]]) -> int:
        return self.self.self.check(grid, 0, 0)
```

```python
from sortedcontainers import SortedSet
from collections import defaultdict

class FoodRatings:

    def __init__(self, foods: List[str], cuisines: List[str], ratings: List[int]):
        self.cuisines_food = defaultdict(SortedSet)
        self.food = {}
        
        for i in range(len(foods)):
            self.cuisines_food[cuisines[i]].add((-ratings[i], foods[i]))
            self.food[foods[i]] = cuisines[i]

    def changeRating(self, food: str, newRating: int) -> None:
        cuisine = self.food[food]
        self.cuisines_food[cuisine].remove()
        self.cuisines_food[cuisine].add((food,newRating))

    def highestRated(self, cuisine: str) -> str:
        return self.cuisines_food[cuisine][0][1]
```
